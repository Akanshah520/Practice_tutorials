```python
import numpy as np
import pandas as pd

# Re-initialize ip DataFrame
ip = pd.read_csv("/content/Coffee_import.csv")

# Re-apply previous transformations
ip.drop(columns = ["1990"], inplace = True)

ip["Buy_again"] = np.where(ip["Total_import"]<830800000, False, True)

ip["Avg_import"] = ip["Total_import"]/30

ip = ip.rename(columns = {"Buy_again":"Repurchased"})

# Introduce NaNs for the first set of indices in '2010'
ip.loc[[0,3,5,7,23,2,6,12,24,30],'2010'] = np.nan

# Fill all NaNs in the DataFrame using the mean of '2010'
ip.fillna(ip['2010'].mean(), inplace =True)

# Reintroduce NaNs for the second set of indices in '2010' (to demonstrate interpolate)
ip.loc[[2,4,7,8,13,15,28], '2010'] = np.nan

# Correctly interpolate the '2010' column
ip['2010'] = ip['2010'].interpolate()

ip['2010'].head()
```
